```python
#training
#1.Load the data (Excel sheet CV)
import pandas as pd
file = pd.read_csv("/content/irisexcel.csv")
x=file[["sepal_length","sepal_width","petal_length","petal_width"]]
y=file["species"]
#cross validation
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
#2.Load the Model(perceptron)
from sklearn.linear_model import Perceptron
model=Perceptron()
#3.Fit the data into the model
model.fit(xtrain,ytrain)
#validation
op=model.predict(xtest)
from sklearn.metrics import accuracy_score
score=accuracy_score(ytest,op)
print("Accuracy=",score*100)
#testing
op2=model.predict([[1.5,2.5,3.5,4.5]])
print(op2)
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(ytest,op)
print(cm)
cmd=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
cmd.plot()
```
